```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame(np.random.rand(100,1), columns=["user_score"])
```

```python
df["id"] = np.arange(100)+1000
df["user_rank"] = df.user_score.rank(method="first",ascending=False).astype(int)
```

```python
attributes = list(df.columns)
attributes.remove("id")
```

```python
attributes
```

```python
res = pd.DataFrame(df[attributes].astype(str).apply(lambda x: x.name+":"+x).apply(lambda x: ";".join(x), axis=1))
```

```python
res["id"] = df.id
```
